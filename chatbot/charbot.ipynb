```python
import gradio as gr
from hugchat import hugchat
from hugchat.login import Login

# Initialize the chatbot (assuming cookies and login are handled)
chatbot = hugchat.ChatBot(cookies=cookies.get_dict())  # Replace {} with actual cookie dictionary if needed

def chat_with_bot(message, history):
    # Fetching the actual response from the chatbot
    response = chatbot.chat(message)
    # Adding a line after each response for better readability
    new_history = history + f"\nUser: {message}\nBot: {response}\n" + "-" * 50 + "\n"
    return response, new_history

# Setup Gradio interface with Blocks for layout control
with gr.Blocks() as app:
    with gr.Row():
        with gr.Column(scale=1):  # Narrower column for history
            history = gr.Textbox(lines=10, label="Chat History", value="Start chatting!", interactive=False, elem_id="chat_history")
        with gr.Column(scale=2):  # Wider column for message and response
            message = gr.Textbox(label="Type your message here", placeholder="Enter your message...")
            response = gr.Textbox(lines=10, label="Chatbot Response", placeholder="Responses appear here", interactive=False)
            submit = gr.Button("Send")

    file_upload = gr.File(label="Upload files")

    submit.click(
        chat_with_bot,
        inputs=[message, history],
        outputs=[response, history]
    )

app.launch()
```
